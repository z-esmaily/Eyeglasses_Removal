```python
def augment_and_save(input_folder, output_folder):
    # Create output folder if it doesn't exist
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    files = os.listdir(input_folder)
    for file_name in files:
        if file_name.lower().endswith(('.png', '.jpg', '.jpeg', '.gif', '.bmp')):
            if not file_name.endswith('mask.png'):

              input_path = os.path.join(input_folder, file_name)
              image = Image.open(input_path)

              # Original flipped image
              save_image(output_folder, file_name, image, '_orig')

              # Original flipped image
              flipped_image = image.transpose(Image.FLIP_LEFT_RIGHT)
              save_image(output_folder, file_name, flipped_image, '_flipped')

def save_image(output_folder, original_name, augmented_image, suffix):
    output_file_name = os.path.splitext(original_name)[0] + suffix + os.path.splitext(original_name)[1]
    output_path = os.path.join(output_folder, output_file_name)
    augmented_image.save(output_path)
```

# **EVALUATE NON GLASSES IMAGES WITH REMOVED GLASSES IMAGES**

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print('Using device:', device)
model = InceptionResnetV1(pretrained='vggface2').eval().to(device)

def extract_features(image_path):
    with torch.no_grad():
        image = Image.open(image_path)
        image_tensor = transform(image).unsqueeze(0)
        features = model(image_tensor).squeeze().numpy()
    return features

# Define the cosine similarity function
def cosine_similarity(a, b):
    dot_product = np.dot(a, b)
    norm_a = np.linalg.norm(a)
    norm_b = np.linalg.norm(b)
    similarity = dot_product / (norm_a * norm_b)
    return similarity


# Define the transforms applied to each image
transform = transforms.Compose([
    transforms.Resize(256),
    transforms.ToTensor(),
    # transforms.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5])
    transforms.Normalize(mean=[0.5, 0.45, 0.40], std=[0.27, 0.23, 0.23]) # good
])

def cal_tar_at_far(gallery_folder, probe_folder, desired_fars):

    gallery_images_list = os.listdir(gallery_folder)
    probe_images_list = os.listdir(probe_folder)

    gallery_features = []
    probe_features = []

    for gallery_image in gallery_images_list:
        if not gallery_image.endswith('mask.png'):
            gallery_image_path = os.path.join(gallery_folder, gallery_image)
            gallery_features.append(extract_features(gallery_image_path))

    for probe_image in probe_images_list:
        if not probe_image.endswith('mask.png'):
            probe_image_path = os.path.join(probe_folder, probe_image)
            probe_features.append(extract_features(probe_image_path))

    gallery_features = np.array(gallery_features)
    probe_features = np.array(probe_features)

    similarities = np.zeros((len(probe_features), len(gallery_features)))
    for i, probe_feature in enumerate(probe_features):
        for j, gallery_feature in enumerate(gallery_features):
            similarity = cosine_similarity(probe_feature, gallery_feature)
            similarities[i, j] = similarity

    y_true = np.zeros(len(probe_images_list) * len(gallery_images_list))
    y_score = np.zeros(len(probe_images_list) * len(gallery_images_list))
    for i, probe_feature in enumerate(probe_features):
        for j, gallery_feature in enumerate(gallery_features):
            index = i * len(gallery_images_list) + j
            y_true[index] = int(gallery_images_list[j].split("@")[:2] == probe_images_list[i].split("@")[:2])
            y_score[index] = similarities[i, j]

    fpr, tpr, thresholds = roc_curve(y_true, y_score)

    # plt.title('Receiver Operating Characteristic')
    # plt.plot(fpr, tpr, 'b')
    # plt.legend(loc = 'lower right')
    # plt.plot([0, 1], [0, 1],'r--')
    # plt.xlim([0, 1])
    # plt.ylim([0, 1])
    # plt.ylabel('True Positive Rate')
    # plt.xlabel('False Positive Rate')
    # plt.show()

    for desired_far in desired_fars:
        # Find the index where FAR is closest to the desired FAR
        index = np.argmin(np.abs(fpr - desired_far))
        tar_at_desired_far = tpr[index]
        print("TAR at FAR =", desired_far, "is", tar_at_desired_far)

    # Calculate accuracy
    y_pred = (y_score >= thresholds[index]).astype(int)
    acc = accuracy_score(y_true, y_pred)
    print("Accuracy:", acc)

    return fpr, tpr, thresholds, y_true, y_score
```

NOTE: The number of samples should be greater than the dimension of the coding layer, here 2048 for the Inception pool 3 layer. Otherwise the covariance is not full rank resulting in complex numbers and nans by calculating the square root. (We doubled the number of test samples so that there is no computational problem due to the number of samples being less than 2048.)

```python
gallery_folder = root_path + rf'/TestDataAndResults/{folder_name}/JustImg/noglasses/gallery'
probe_folder = root_path + rf'/TestDataAndResults/{folder_name}/JustImg/removed_by_default/probe_double'
desired_fars = [0.1, 0.01, 0.001]
cal_tar_at_far(gallery_folder, probe_folder, desired_fars)

gallery_folder = root_path + rf'/TestDataAndResults/{folder_name}/JustImg/noglasses/gallery'
probe_folder = root_path + rf'/TestDataAndResults/{folder_name}/JustImg/removed_by_prop/probe_double'
desired_fars = [0.1, 0.01, 0.001]
cal_tar_at_far(gallery_folder, probe_folder, desired_fars)

gallery_folder = root_path + rf'/TestDataAndResults/{folder_name}/JustImg/noglasses/gallery'
probe_folder = root_path + rf'/TestDataAndResults/{folder_name}/JustImg/removed_by_HiSD/probe'
desired_far = [0.1, 0.01, 0.001]
cal_tar_at_far(gallery_folder, probe_folder, desired_far)

gallery_folder = root_path + rf'/TestDataAndResults/{folder_name}/JustImg/noglasses/gallery'
probe_folder = root_path + rf'/TestDataAndResults/{folder_name}/JustImg/removed_by_SAGAN'
desired_far = [0.1, 0.01, 0.001]
cal_tar_at_far(gallery_folder, probe_folder, desired_far)

gallery_folder = root_path + rf'/TestDataAndResults/{folder_name}/JustImg/noglasses/gallery'
probe_folder = root_path + rf'/TestDataAndResults/{folder_name}/JustImg/removed_by_ERGAN120*120'
desired_far = [0.1, 0.01, 0.001]
cal_tar_at_far(gallery_folder, probe_folder, desired_far)
```
