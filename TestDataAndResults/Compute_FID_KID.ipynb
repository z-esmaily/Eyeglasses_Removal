# **CREATE ONLY IMAGE FOLDERS**

```python
%cd '/content/drive/MyDrive/Eyeglasses_Removal-main/'
```

```python
def copy_only_imges(src_path, dst_path):
  for filename in os.listdir(src_path):
    if not filename.endswith('mask.png'):
      shutil.copy(src_path + filename, dst_path + filename)

  print('check', len(os.listdir(src_path)), len(os.listdir(dst_path)))
```

```python
path1 = './TestDataAndResults/CelebA/removed_by_prop/'
path2 = './TestDataAndResults/CelebA/JustImg/removed_by_prop/'
if not os.path.exists(path2):
    os.makedirs(path2)
copy_only_imges(path1, path2)

print('All portrait images of: ', path1[:-6], '\n copied to: ', path2)
```
