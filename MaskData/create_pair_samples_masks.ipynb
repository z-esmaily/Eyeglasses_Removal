```python
import os
import cv2 as cv2
import numpy as np


# training folders
before_thk7_folder = './MaskData/train_samples/before_thk7/'
before_thk5_folder = './MaskData/train_samples/before_thk5/'

no_damage_folder = './MaskData/train_samples/no_damage/'

after_thk7_folder = './MaskData/train_samples/after_thk7/'
after_thk5_folder = './MaskData/train_samples/after_thk5/'

# validation folders
val_before_thk7_folder = './MaskData/validation_samples/before_thk7/'
val_before_thk5_folder = './MaskData/validation_samples/before_thk5/'

val_no_damage_folder = './MaskData/validation_samples/no_damage/'

val_after_thk7_folder = './MaskData/validation_samples/after_thk7/'
val_after_thk5_folder = './MaskData/validation_samples/after_thk5/'
```

# **TRAIN SET**

```python
before_thk7_list = os.listdir(before_thk7_folder)
before_thk5_list = os.listdir(before_thk5_folder)
no_damage_list = os.listdir(no_damage_folder)
print(len(before_thk7_list), len(before_thk5_list), len(no_damage_list))
```

# **Fill source training folders (after folders)**

```python
def tophat7(img):
  kernel = np.ones((7,7),np.uint8)
  tophat = cv2.morphologyEx(img, cv2.MORPH_TOPHAT, kernel)
  return tophat


def tophat5(img):
  kernel = np.ones((5,5),np.uint8)
  tophat = cv2.morphologyEx(img, cv2.MORPH_TOPHAT, kernel)
  return tophat


def load_images_apply_tophat(in_folder, out_folder, before_list, thk):
  images = []
  for filename in before_list:
    # print(filename)
    img = cv2.imread(os.path.join(in_folder, filename), 0)
    if thk == 'thk7':
      tophat_image7 = tophat7(img)
      cv2.imwrite(out_folder + filename, tophat_image7)
    elif thk == 'thk5':
      tophat_image5 = tophat5(img)
      cv2.imwrite(out_folder + filename, tophat_image5)
    else:
      print("thk is Not defined")
  return images

load_images_apply_tophat(before_thk7_folder, after_thk7_folder, before_thk7_list, 'thk7')
load_images_apply_tophat(before_thk5_folder, after_thk5_folder, before_thk5_list, 'thk5')
```

```python
after_thk7_list = os.listdir(after_thk7_folder)
after_thk5_list = os.listdir(after_thk5_folder)
no_damage_list = os.listdir(no_damage_folder)
print(len(after_thk7_list), len(after_thk5_list), len(no_damage_list))
```

# **VALIDATION SET**

```python
val_before_thk7_list = os.listdir(val_before_thk7_folder)
val_before_thk5_list = os.listdir(val_before_thk5_folder)
val_no_damage_list = os.listdir(val_no_damage_folder)
print(len(val_before_thk7_list), len(val_before_thk5_list), len(val_no_damage_list))
```

# **Fill source validation folders (after folders)**

```python
load_images_apply_tophat(val_before_thk7_folder, val_after_thk7_folder, val_before_thk7_list, 'thk7')
load_images_apply_tophat(val_before_thk5_folder, val_after_thk5_folder, val_before_thk5_list, 'thk5')
```

```python
val_after_thk7_list = os.listdir(val_after_thk7_folder)
val_after_thk5_list = os.listdir(val_after_thk5_folder)
val_no_damage_list = os.listdir(val_no_damage_folder)
print(len(val_after_thk7_list), len(val_after_thk5_list), len(val_no_damage_list))
```
